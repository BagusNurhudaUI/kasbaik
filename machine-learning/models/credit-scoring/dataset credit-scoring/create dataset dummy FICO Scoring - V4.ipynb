```python
data  = pd.read_csv(r'./Dummy Datasets FICO Scoring - V4 - Belum Kategori.csv', delimiter=',')
print(data.head())
```

```python
def Cat_to_Num(features):
    for feature in features:
        feature_list = list(np.unique(data[feature]))
        feature_dict = {}
        for i in range(len(feature_list)):
                       feature_dict[feature_list[i]] = i
        data.replace({feature : feature_dict}, inplace=True)
        print(feature, '-->', feature_dict)
```

```python
categorical_features = ['pekerjaan']
Cat_to_Num(categorical_features)
```

```python
# converting to list
x1 = data["usia"].tolist()
x2 = data["pinjaman"].tolist()
x3 = data["tenor"].tolist()
x4 = data["pemasukan"].tolist()
x5 = data["tanggungan"].tolist()
x6 = data["pekerjaan"].tolist()
x7 = data["pinjaman ke-"].tolist()
x8 = data["telat bayar"].tolist()
x9 = data["donasi"].tolist()
```

```python
# X10 - Kondisi Ekonomi (econ)
x10 = []
for i in range(0,10000):
# for i in range(0,10):
    x = x4[i] - ((x2[i]/(x3[i]/4)) +150000*(x5[i]+2))
    if x >= 2000000:
        x10.append(5)
    elif 1500000 <= x < 2000000:
        x10.append(4)
    elif 1000000 <= x < 1500000:
        x10.append(3)
    elif 500000 <= x < 1000000:
        x10.append(2)
    elif 100000 <= x < 500000:
        x10.append(1)
    else:
        x10.append(0)
print(len(x10))
        
```

```python
# X11 - Usia
x11 = []
for i in range(0,10000):
# for i in range(0,10):
    if x1[i] > 64:
        x11.append(1)
    else:
        x11.append(2)
print(len(x11))
```

```python
# X12 - Pekerjaan
x12 = []
for i in range(0,10000):
# for i in range(0,10):
    if x6[i]  == 0 or x6[i]  == 1 :
        x12.append(2)
    else:
        x12.append(1)
print(len(x12))
```

```python
# X13 - Telat Bayar
x13 = []
for i in range(0,10000):
# for i in range(0,10):
    if x8[i] == 0:
        x13.append(0)
    elif x8[i] == 1:
        x13.append(1)
    elif 2 <= x8[i] <= 4:
        x13.append(2)
    else:
        x13.append(3)
print(len(x13))
```

```python
# X14 - Pinjaman Ke
x14 = []
for i in range(0,10000):
# for i in range(0,10):
    if x7[i] == 1:
        x14.append(1)
    elif x7[i] == 2 or x7[i] == 3 :
        x14.append(2)
    else:
        x14.append(3)
print(len(x14))
```

```python
# X15 - Donasi
x15 = []
for i in range(0,10000):
# for i in range(0,10):
    if x9[i] == 0:
        x15.append(0)
    elif x9[i] >= 1 and x9[i] <= 3 :
        x15.append(1)
    elif x9[i] > 3 and x9[i] <= 5 :
        x15.append(2)
    else:
        x15.append(3)
print(max(x15))
```
